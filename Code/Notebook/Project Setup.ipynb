```
# Import libaries

from matplotlib import pyplot as plt
from matplotlib.patches import Patch
import matplotlib.colors as mcolors
import pandas as pd
import os
import glob  
import geopandas as gpd
```

```
# Import Shapefiles

### Tracts
T_shp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Tracts/nyct2020.shp"
T_gdf_shapes = gpd.read_file(T_shp_path)

### Districts
D_shp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Districts/nycd.shp"
D_gdf_shapes = gpd.read_file(D_shp_path)

### Neighborhoods
N_shp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Neighborhoods/nynta2020.shp"
N_gdf_shapes = gpd.read_file(N_shp_path)
```

```
# Map out the shapefiles

### Convert all shapefiles to the same CRS (EPSG:4326) (the standard NYC Census Tract CRS)
T_gdf_shapes = T_gdf_shapes.to_crs("EPSG:4326")
D_gdf_shapes = D_gdf_shapes.to_crs(T_gdf_shapes.crs)
N_gdf_shapes = N_gdf_shapes.to_crs(T_gdf_shapes.crs)

### Create the figure and axes
fig, ax = plt.subplots(figsize=(7, 7))

### Layer 1: Census Tracts
T_gdf_shapes.plot(ax=ax, color='lightgray', edgecolor='green', linewidth=0.5, alpha=0.5, zorder=1)

### Layer 2: Neighborhoods
N_gdf_shapes.plot(ax=ax, color='none', edgecolor='blue', linewidth=1, zorder=2)

### Layer 3: Districts
D_gdf_shapes.plot(ax=ax, color='none', edgecolor='lightcoral', linewidth=1, zorder=3)

### Formatting
plt.title("NYC Overlap: Districts, Neighborhoods, Tracts")
ax.axis('off')
plt.show()
```
