```
# Import libaries

from matplotlib import pyplot as plt
from matplotlib.patches import Patch
import matplotlib.colors as mcolors
import pandas as pd
import os
import glob  
import geopandas as gpd
```

```
# Import Datasets

### Census Tracts 
csv_path = ("/work/AP_RES_PRO/Data/Processed/Census_Tracts/2020_Census_Tracts_20251210.csv")
df = pd.read_csv(csv_path)

# Import Shapefiles

### Tracts
Tshp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Tracts/nyct2020.shp"
Tgdf_shapes = gpd.read_file(Tshp_path)

### Districts
Dshp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Districts/Districts2022-10-06.shp"
Dgdf_shapes = gpd.read_file(Dshp_path)

### Neighborhoods
Nshp_path = "/work/AP_RES_PRO/Data/Processed/Shapefiles/Neighborhoods/geo_export_b142f009-acf1-4d1c-aa9b-49e7bd3bffd6.shp"
Ngdf_shapes = gpd.read_file(Nshp_path)
```

```
# Convert all shapefiles to the same CRS (EPSG:4326) (the standard NYC Census Tract CRS)
Tgdf_shapes = Tgdf_shapes.to_crs("EPSG:4326")
Dgdf_shapes = Dgdf_shapes.to_crs(Tgdf_shapes.crs)
Ngdf_shapes = Ngdf_shapes.to_crs(Tgdf_shapes.crs)

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 10))

# Layer 1: Census Tracts
Tgdf_shapes.plot(ax=ax, color='lightgray', edgecolor='green', linewidth=0.5, alpha=0.5, zorder=1)

# Layer 2: Neighborhoods
Ngdf_shapes.plot(ax=ax, color='none', edgecolor='blue', linewidth=1, zorder=2)

# Layer 3: Districts
Dgdf_shapes.plot(ax=ax, color='none', edgecolor='lightcoral', linewidth=1, zorder=3)

# Formatting
plt.title("NYC Overlap: Districts, Neighborhoods, Tracts")
ax.axis('off')
plt.show()
```
